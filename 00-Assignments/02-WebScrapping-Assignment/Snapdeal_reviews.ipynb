```python
from bs4 import BeautifulSoup as bs
import requests

link = 'https://www.snapdeal.com/product/campus-crunch-blue-running-shoes/629787596130/reviews?page=2&sortBy=HELPFUL'

headers = {"User-Agent": "Mozilla/5.0"}

page = requests.get(link, headers=headers)

soup = bs(page.content, 'html.parser')

print(soup.prettify())
```

```python
# Find all user reviews
reviews = soup.find_all('div', class_='user-review')

for review in reviews:

    rating = len(review.find_all('i', class_='sd-icon sd-icon-star active'))


    review_title_div = review.find('div', class_='head')
    review_title = review_title_div.get_text(strip=True) if review_title_div else 'No title found'


    reviewer_info_div = review.find('div', class_='_reviewUserName')
    reviewer_info = reviewer_info_div.get_text(strip=True) if reviewer_info_div else 'No reviewer info found'

    review_body_p = review.find('p')
    review_body = review_body_p.get_text(strip=True) if review_body_p else 'No review body found'

    # Print extracted information
    print(f"Rating: {rating} stars")
    print(f"Title: {review_title}")
    print(f"Reviewer: {reviewer_info}")
    print(f"Review: {review_body}")
    print('-' * 50)
```
