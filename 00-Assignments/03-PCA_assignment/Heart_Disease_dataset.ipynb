### **Business Objective:**
The primary objective for the pharmaceutical company is to analyze patient data to better understand patterns related to heart disease. Specifically, the goal is to:
- **Segregate patients** into different groups based on various factors like age and other medical attributes in the dataset.
- This analysis will help the company gain insights into how different groups of patients might respond to the new heart disease treatment or exhibit different risk profiles.

The analysis involves applying dimensionality reduction techniques (Principal Component Analysis, PCA) and clustering algorithms to identify meaningful patterns and relationships between different patient attributes. By comparing the clustering results before and after PCA, the company can understand how dimensionality reduction impacts the grouping of patients and if better, more interpretable clusters can be formed. This would provide key insights that may help guide treatment plans, research, or drug development strategies.

### **Constraints:**
1. **Data Quality**:
   - The data is sourced from secondary sources, which could have issues like missing values, inconsistent formats, or outliers that need to be handled before analysis.

2. **Dimensionality**:
   - The dataset may have many features, some of which could be irrelevant or highly correlated. Reducing dimensionality through PCA can help make the clustering more efficient, but care must be taken not to lose important information.

3. **Model Interpretation**:
   - After applying PCA and clustering, the results must be interpretable by business stakeholders. It’s essential to explain the differences between the clusters formed before and after PCA in a way that’s meaningful for the decision-making process.

4. **Scalability**:
   - Depending on the size of the dataset, the analysis should be computationally efficient. High-dimensional data can increase computational time, especially for clustering algorithms like K-means and hierarchical clustering.

5. **Patient Privacy**:
   - Since the data involves patients, there may be privacy and security constraints in handling the data, requiring anonymization.

6. **Cluster Validity**:
   - Validating the clusters and ensuring that they represent meaningful groups is critical. The company would want insights that reflect actual patient segments and not arbitrary groupings driven by noise in the data.

```python
#kmeans clustering on the scaled and cleaned data
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#plot age vs trestbps clusters along with the centers
kmeans.fit(df[['age', 'trestbps']])
#create different dataset for each cluster
cluster_1 = df[kmeans.labels_ == 0]
cluster_2 = df[kmeans.labels_ == 1]
cluster_3 = df[kmeans.labels_ == 2]
#cluster centers
cluster_centers = kmeans.cluster_centers_

plt.figure(figsize=(10, 6))
plt.scatter(cluster_1['age'], cluster_1['trestbps'], label='Cluster 1', color='blue')
plt.scatter(cluster_2['age'], cluster_2['trestbps'], label='Cluster 2', color='green')
plt.scatter(cluster_3['age'], cluster_3['trestbps'], label='Cluster 3', color='red')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='*', s=200, color='black', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.title('Clusters of Age vs Resting Blood Pressure')
plt.legend()
plt.show()
```

```python
#Normalize the data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)
df_scaled=pd.DataFrame(scaled_data,columns=df.columns)
```

```python
columns_with_outliers=['trestbps','chol','thalach','oldpeak','ca','thal','fbs']
```

```python
# prompt: apply log transformation on columns with outliers

import numpy as np

for column in df.columns:
  df_scaled[column] = np.log1p(df_scaled[column])

# Boxplot after log transformation
plt.figure(figsize=(15, 10))
sns.boxplot(data=df_scaled)
plt.xticks(rotation=90)
plt.title('Boxplot of Scaled Heart Disease Data after Log Transformation')
plt.show()
```

```python
#remove the outliers
def remove_outliers(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        #Remove outliers
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

df_cleaned = remove_outliers(df_scaled, columns_with_outliers)
```

```python
#kmeans clustering on the scaled and cleaned data
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
#plot age vs trestbps clusters along with the centers
y_kmeans=kmeans.fit_predict(df_cleaned[['age', 'trestbps']])
#create different dataset for each cluster
cluster_1 = df_cleaned[kmeans.labels_ == 0]
cluster_2 = df_cleaned[kmeans.labels_ == 1]
cluster_3 = df_cleaned[kmeans.labels_ == 2]
#cluster centers
cluster_centers = kmeans.cluster_centers_

plt.figure(figsize=(10, 6))
plt.scatter(cluster_1['age'], cluster_1['trestbps'], label='Cluster 1', color='blue')
plt.scatter(cluster_2['age'], cluster_2['trestbps'], label='Cluster 2', color='green')
plt.scatter(cluster_3['age'], cluster_3['trestbps'], label='Cluster 3', color='red')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], marker='X', s=200, color='black', label='Centroids')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.title('Clusters of Age vs Resting Blood Pressure')
plt.legend()
plt.show()
```

```python
#calculate silhoutte score
from sklearn.metrics import silhouette_score
sil_score1=silhouette_score(df_cleaned,y_kmeans)
print(sil_score1)
```

# Principal component analysis

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
principal_components = pca.fit_transform(df)
# Create a new DataFrame with the principal components
principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2', 'PC3'])
# Print the explained variance ratio for each component
print('Explained Variance Ratio:', pca.explained_variance_ratio_)
```

```python
#perform kmeans clustering on PC1 and PC2
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(principal_df[['PC1', 'PC2']])
# Get cluster centers (centroids)
centroids = kmeans.cluster_centers_
# Create dataframes for different clusters
df_cluster1 = principal_df[y_kmeans == 0]
df_cluster2 = principal_df[y_kmeans == 1]
df_cluster3 = principal_df[y_kmeans == 2]
# Plot the clusters and centroids
plt.scatter(df_cluster1['PC1'], df_cluster1['PC2'], s=50, c='blue', label='Cluster 1')
plt.scatter(df_cluster2['PC1'], df_cluster2['PC2'], s=50, c='green', label='Cluster 2')
plt.scatter(df_cluster3['PC1'], df_cluster3['PC2'], s=50, c='red', label='Cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='black', label='Centroids')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('K-means Clustering of Wine Dataset (PC1 vs PC2)')
plt.legend()
plt.show()
```

```python
#calculate silhoutte score
sil_score_pca1=silhouette_score(principal_df,y_kmeans)
print(sil_score_pca1)
```

```python
#perform kmeans clustering on PC1 and PC3
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(principal_df[['PC1', 'PC3']])
# Get cluster centers (centroids)
centroids = kmeans.cluster_centers_
# Create dataframes for different clusters
df_cluster1 = principal_df[y_kmeans == 0]
df_cluster2 = principal_df[y_kmeans == 1]
df_cluster3 = principal_df[y_kmeans == 2]
# Plot the clusters and centroids
plt.scatter(df_cluster1['PC1'], df_cluster1['PC3'], s=50, c='blue', label='Cluster 1')
plt.scatter(df_cluster2['PC1'], df_cluster2['PC3'], s=50, c='green', label='Cluster 2')
plt.scatter(df_cluster3['PC1'], df_cluster3['PC3'], s=50, c='red', label='Cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='black', label='Centroids')
plt.xlabel('PC1')
plt.ylabel('PC3')
plt.title('K-means Clustering of Wine Dataset (PC1 vs PC3)')
plt.legend()
plt.show()
```

```python
#calculate silhoutte score
sil_score_pca2=silhouette_score(principal_df,y_kmeans)
print(sil_score_pca2)
```

```python
#perform kmeans clustering on PC2 and PC3
kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(principal_df[['PC2', 'PC3']])
# Get cluster centers (centroids)
centroids = kmeans.cluster_centers_
# Create dataframes for different clusters
df_cluster1 = principal_df[y_kmeans == 0]
df_cluster2 = principal_df[y_kmeans == 1]
df_cluster3 = principal_df[y_kmeans == 2]
# Plot the clusters and centroids
plt.scatter(df_cluster1['PC2'], df_cluster1['PC3'], s=50, c='blue', label='Cluster 1')
plt.scatter(df_cluster2['PC2'], df_cluster2['PC3'], s=50, c='green', label='Cluster 2')
plt.scatter(df_cluster3['PC2'], df_cluster3['PC3'], s=50, c='red', label='Cluster 3')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=200, c='black', label='Centroids')
plt.xlabel('PC2')
plt.ylabel('PC3')
plt.title('K-means Clustering of Wine Dataset (PC2 vs PC3)')
plt.legend()
plt.show()
```

```python
#calculate silhoutte score
sil_score_pca3=silhouette_score(principal_df,y_kmeans)
print(sil_score_pca3)
```

```python
#Average Silhoutte score for principal component clustering
avg=(sil_score_pca1+sil_score_pca2+sil_score_pca3)/3
avg
```
